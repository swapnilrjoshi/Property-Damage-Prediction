```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
from sklearn.model_selection import train_test_split
from xgboost.sklearn import XGBClassifier
from xgboost import plot_importance
import shap
from KFoldEncoder import KFoldTargetEncoderTrain, KFoldTargetEncoderTest
from custom_functions import magnitude_feature, modelfit, get_modelReport
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('weather-alerts-challenge.csv')
data.head(5)
```

```python
#Converting Windspeed to numerical data
data['Windspeed'] = data['Windspeed'].apply(lambda x: float(x.split()[0]))
```

```python
#Creating categorical features for Magnitude 
data['Magnitude']=data['Magnitude'].fillna('')
data['HailMag'],data['WindMag'],data['NonMagEvent']=zip(*data['Magnitude'].map(magnitude_feature))

#Data Binning
hail_bins = [0, 0.01, 0.75, 0.88, 1.25, 2.75, 5.5]
hail_labels = [0,1,2,3,4,5]
data['HailMag_bin'] = pd.cut(data['HailMag'], bins=hail_bins, labels=hail_labels,include_lowest =True)
```

```python
#Creating feature for Episode Narrative len
data['Episode Narrative len'] = data['Episode Narrative'].apply(len) 
```

```python
#Spliting Training and Testing Data
train, test = train_test_split(data, test_size=0.25, random_state=4, stratify=data['IsPropertyDamage'])
```

```python
##Defining categorical variables
cat_cols=['Event','Report Source','HailMag_bin']

targetc = KFoldTargetEncoderTrain(cat_cols,'IsPropertyDamage',n_fold=5)
train,train_map = targetc.fit_transform(train)

test_targetc = KFoldTargetEncoderTest(train_map,cat_cols)
test=test_targetc.fit_transform(test)
```

```python
col_list=list(train.columns)
train_list=[12,13,16,19,20,21,22]
train_cols=[]
for i in train_list:
    train_cols.append(col_list[i])

X_train=train[train_cols]
y_train=train['IsPropertyDamage'] 

X_test=test[train_cols]
y_test=test['IsPropertyDamage']
```

```python
X_train.head(5)
```

# Model

```python
xgb = XGBClassifier(learning_rate =0.05,n_estimators=280,max_depth=5,
                     min_child_weight=1,gamma=0.4,subsample=0.8,colsample_bytree=0.8,
                     objective= 'binary:logistic', nthread=4, scale_pos_weight=2.75, seed=28, max_delta_step=3.0)
xgb = modelfit(xgb, X_train, y_train)
```

```python
xgb
```

```python
#Training data predictions
ypred_train = xgb.predict(X_train)
yproba_train = xgb.predict_proba(X_train)
get_modelReport(y_train,ypred_train,yproba_train)
```

```python
#Test data predictions
ypred_test = xgb.predict(X_test)
yproba_test = xgb.predict_proba(X_test)
get_modelReport(y_test,ypred_test,yproba_test)
```

```python
#Changing Threshold
#Train predictions after changing threshold
threshold = 0.35
newpred_train = pd.Series(np.where(yproba_train[:,1] > threshold, 1, 0))
get_modelReport(y_train,newpred_train,yproba_train)
```

```python
#Test predictions after threshold
newpred_test = pd.Series(np.where(yproba_test[:,1] > threshold, 1, 0))
get_modelReport(y_test,newpred_test,yproba_test)
```

```python
#Plotting Feature Importance
plot_importance(xgb)               
plot_importance(xgb,importance_type='gain')               
plot_importance(xgb,importance_type='cover')    
```

# Train Final Model on Full Data

```python
targetc = KFoldTargetEncoderTrain(cat_cols,'IsPropertyDamage',n_fold=5)
full_data,X_map = targetc.fit_transform(data)
X = full_data[train_cols]
y= full_data['IsPropertyDamage']
X.head(5)
```

```python
xgb = XGBClassifier(learning_rate =0.05,n_estimators=280,max_depth=5,
                     min_child_weight=1,gamma=0.4,subsample=0.8,colsample_bytree=0.8,
                     objective= 'binary:logistic', nthread=4, scale_pos_weight=2.75, seed=28,max_delta_step=3.0)
xgb = modelfit(xgb, X, y)
xgb
```

```python
#Data predictions
ypred = xgb.predict(X)
yproba = xgb.predict_proba(X)
get_modelReport(y,ypred,yproba)
```

```python
#Changing Threshold
#Predictions after changing threshold
threshold = 0.35
newpred = pd.Series(np.where(yproba[:,1] > threshold, 1, 0))
get_modelReport(y,newpred,yproba)
```
